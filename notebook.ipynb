## Jak to wygląda w praktyce?

- **Efektywność energetyczna** mierzy, ile energii potrzeba do uzyskania określonego efektu (np. wyprodukowania jednostki PKB, ogrzania budynku, transportu tony towaru). Im mniej energii zużywasz na ten sam efekt, tym wyższa efektywność.
- **OZE** same w sobie nie zawsze bezpośrednio zwiększają efektywność energetyczną, bo efektywność dotyczy raczej tego, jak dobrze wykorzystywana jest energia, niezależnie od jej źródła. Jednak przechodzenie na OZE często wiąże się z modernizacją infrastruktury, lepszymi technologiami i bardziej efektywnym zużyciem energii, co może pośrednio poprawiać efektywność energetyczną[1][2][3].

## Co pokazują dane Eurostatu?

- Statystyki Eurostatu pozwalają śledzić zarówno udział OZE w zużyciu energii, jak i wskaźniki efektywności energetycznej (np. energochłonność gospodarki, zużycie energii końcowej na jednostkę PKB).
- Możesz porównywać kraje lub lata pod kątem tego, czy wzrost udziału OZE idzie w parze ze spadkiem energochłonności lub zużycia energii na mieszkańca[2][3].
- Badania naukowe (np. [1]) pokazują, że wzrost udziału OZE w miksie energetycznym UE koreluje z poprawą wskaźników środowiskowych (np. spadkiem emisji CO₂), a często także z poprawą efektywności energetycznej, choć wpływ ten jest pośredni i zależy od wielu czynników (technologie, struktura gospodarki, polityka energetyczna).

## Na co zwrócić uwagę?

- Przykładowo, jeśli kraj zwiększa udział OZE i jednocześnie spada jego zużycie energii na PKB, można mówić o pozytywnym wpływie transformacji na efektywność energetyczną.

Różnica między **final energy consumption** (zużyciem energii końcowej) a **primary energy consumption** (zużyciem energii pierwotnej) w raporcie Eurostatu „Energy efficiency” polega na tym, na jakim etapie „drogi” energii przez gospodarkę są one mierzone i co obejmują.

---

## Primary energy consumption (PEC) – zużycie energii pierwotnej

- **Definicja:** To całkowite zapotrzebowanie na energię w kraju, obejmujące:
  - energię zużytą przez końcowych odbiorców,
  - straty podczas produkcji i przesyłu,
  - zużycie energii przez sam sektor energetyczny (np. praca elektrowni),
  - energię zużytą na transformację (np. spalanie węgla do produkcji prądu),
  - z wyłączeniem energii użytej do celów nieenergetycznych (np. ropa do produkcji plastiku)[2][6].
- **Co mierzy:** Pokazuje, ile energii pierwotnie „weszło” do systemu energetycznego kraju — zanim została przetworzona, przesłana i zużyta przez odbiorców końcowych.
- **Przykład:** Jeśli gospodarstwo domowe zużyje 1 kWh prądu, a wyprodukowanie tej 1 kWh w elektrowni węglowej wymagało spalenia węgla o wartości 2,5 kWh energii pierwotnej (ze względu na sprawność i straty), to do PEC zalicza się właśnie te 2,5 kWh.

---

## Final energy consumption (FEC) – zużycie energii końcowej

- **Definicja:** To ilość energii, która dociera do końcowych odbiorców (gospodarstw domowych, przemysłu, transportu, rolnictwa itd.) i jest przez nich zużywana[1][4][6].
- **Co mierzy:** Pokazuje realne zużycie energii „na miejscu” – to, co faktycznie trafia do użytkowników po wszystkich transformacjach i przesyle.
- **Przykład:** Jeśli gospodarstwo domowe zużyje 1 kWh prądu, to właśnie ta 1 kWh wlicza się do FEC.

---

## Różnice i zależności

- **PEC jest zawsze większe od FEC**, ponieważ obejmuje straty, zużycie energii przez sektor energetyczny i procesy transformacji, zanim energia dotrze do końcowego odbiorcy[1][2][6].
- **FEC** to energia „u klienta”, **PEC** to energia „na wejściu” do całej gospodarki.

---

### Podsumowanie w tabeli

| Wskaźnik                    | Co obejmuje?                                                                                   | Przykład                          |
|-----------------------------|-----------------------------------------------------------------------------------------------|-----------------------------------|

| Final energy consumption     | Energia zużyta przez odbiorców końcowych (gospodarstwa, przemysł, transport)                  | 1 kWh prądu zużyte w domu         |

**W skrócie:**  
- **Final energy consumption** – ile energii faktycznie zużywają odbiorcy końcowi.  
- **Primary energy consumption** – ile energii pierwotnej musiało być dostarczone do systemu, aby pokryć całe krajowe zapotrzebowanie, łącznie ze stratami i transformacją[1][2][4][6].
